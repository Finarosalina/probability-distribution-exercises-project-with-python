# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import binom

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """ 
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python

import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

## Ejercicio 1

Escriba las funciones anteriores pero para distribuciones normales y uniformes.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform

# Funciones para la distribución normal

def dnorm(x, mean=0, std=1):  # Cálculo densidad de probabilidad
    result = norm.pdf(x, loc=mean, scale=std)
    return result

def pnorm(q, mean=0, std=1):  # Cálculo acumulado de la distribución normal.
    result = norm.cdf(q, loc=mean, scale=std)
    return result

def qnorm(p, mean=0, std=1):  # Cálculo cuantil de la distribución normal.
    result = norm.ppf(p, loc=mean, scale=std)
    return result

def rnorm(n, mean=0, std=1):  # Genera variables aleatorias a partir de la distribución normal.
    result = norm.rvs(loc=mean, scale=std, size=n)
    return result

# Funciones para la distribución uniforme

def dunif(x, loc=0, scale=1):  # Calcula la densidad de probabilidad (estimación puntual) de la distribución uniforme.
    result = uniform.pdf(x, loc=loc, scale=scale)
    return result

def punif(q, loc=0, scale=1):  # Calcula el acumulado de la distribución uniforme.
    result = uniform.cdf(q, loc=loc, scale=scale)
    return result

def qunif(p, loc=0, scale=1):  # Calcula el cuantil de la distribución uniforme.
    result = uniform.ppf(p, loc=loc, scale=scale)
    return result

def runif(n, loc=0, scale=1):  # Genera variables aleatorias a partir de la distribución uniforme.
    result = uniform.rvs(loc=loc, scale=scale, size=n)
    return result

# Ejemplo de uso con distribuciones normales y uniformes

# Distribución normal(0, 1), donde media = 0, desviación estándar = 1
print(f"Density at 0 for Normal(0, 1): {dnorm(0, mean=0, std=1)}")
print(f"Cumulative probability of 1 for Normal(0, 1): {pnorm(1, mean=0, std=1)}")
print(f"Quantile at 0.95 for Normal(0, 1): {qnorm(0.95, mean=0, std=1)}")
print(f"Generate 5 random values from Normal(0, 1): {rnorm(5, mean=0, std=1)}")

# Distribución uniforme(0, 1), donde loc = 0 y scale = 1
print(f"Density at 0.5 for Uniform(0, 1): {dunif(0.5, loc=0, scale=1)}")
print(f"Cumulative probability of 0.5 for Uniform(0, 1): {punif(0.5, loc=0, scale=1)}")
print(f"Quantile at 0.95 for Uniform(0, 1): {qunif(0.95, loc=0, scale=1)}")
print(f"Generate 5 random values from Uniform(0, 1): {runif(5, loc=0, scale=1)}")

# Grafico distribución normal

np.random.seed(42)

data = rnorm(2000, mean=0, std=1)

pmf = dnorm(data, mean=0, std=1)  # Densidad de probabilidad (solo continuas, pero es un lanzamiento de moneda..)
cdf = pnorm(data, mean=0, std=1)  # Función acumulada
ppf = qnorm(cdf, mean=0, std=1)  # Cuantiles

fig, axis = plt.subplots(3, 1, figsize=(5, 7))

axis[0].stem(data[:50], pmf[:50], "r-")  # Limitar el número de puntos 50
axis[1].stem(data[:50], cdf[:50], "b-")
axis[2].stem(data[:50], ppf[:50], "g-")


plt.tight_layout()

plt.show()


# Grafico distribución uniforme

np.random.seed(42)

# Cambié los argumentos de runif, ahora se usan 'n', 'loc', y 'scale'
data = runif(2000, loc=0, scale=1)

pmf = dunif(data, loc=0, scale=1)
cdf = punif(data, loc=0, scale=1)  # Función acumulada
ppf = qunif(cdf, loc=0, scale=1)  # Cuantiles

print("Distribucion uniforme")
# Graficar
fig, axis = plt.subplots(3, 1, figsize=(5, 7))

axis[0].stem(data[:50], pmf[:50], "r-")  # Limitar el número de puntos a 50
axis[1].stem(data[:50], cdf[:50], "b-")
axis[2].stem(data[:50], ppf[:50], "g-")

print("\n")     

plt.tight_layout()

plt.show()


```
